```python
# 公司数据集
mse = [0.437, 0.294, 0.284, 0.280, 0.291, 0.295, 0.310]
mae = [0.322, 0.278, 0.275, 0.273, 0.276, 0.279, 0.284]
rmse = [0.661, 0.542, 0.533, 0.529, 0.539, 0.548, 0.551]
# index = np.array([1.00, 5.00, 10.00, 15.00, 20.00, 25.00, 30.00])
index = np.array([24, 48, 96, 144, 168, 196, 240])

# Plot setup
plt.figure(facecolor='white', figsize=(10, 6))  # Adjusted for a more standard size
bar_width = 0
# 设置网格线格式：
plt.grid(True)
plt.style.use('classic')
# Creating bars
# plt.bar(index, mse, color='green', width=bar_width, label='MSE')
# plt.bar(index + bar_width, mae, color='blue', width=bar_width, label='MAE')
# plt.bar(index + 2 * bar_width, rmse, color='yellow', width=bar_width, label='RMSE')

plt.plot(index, mse, label='MSE', marker='o')
plt.plot(index, mae, label='MAE', marker='x')
plt.plot(index, rmse, label='RMSE', marker='s')

# Adding legend and labels
plt.legend(loc='upper right', fontsize=22)
plt.xticks(index + bar_width, ['24', '48', '96', '144', '168', '196', '240'])

# Display the plot
plt.tight_layout()  # Adjust layout to not cut off labels
plt.savefig('./' + 'pre_len_influence.svg', format='svg', dpi=1000,
            bbox_inches='tight')
plt.show()
```
